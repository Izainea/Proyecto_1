```python
### Un análisis de correspondencia pra las elecciones de Francia 2017

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prince import CA
from prince import datasets

# Cargamos los datos

elecciones= datasets.load_french_elections()
elecciones
```

```python
### Análisis de correspondencia

# Creamos un objeto de la clase CA

analisis_correspondencia =CA(
    n_components=2,
    n_iter=3,
    copy=True,
    check_input=True,
    engine='sklearn',
    random_state=42
)

ca = analisis_correspondencia.fit(elecciones)
```

```python
ca.eigenvalues_summary
```
